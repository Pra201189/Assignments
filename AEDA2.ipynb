```python
#Handling Missing Values
#Removing rows and columns with too many missing values
df_clean=df.dropna()
df
```

```python
#Applying scaling techniques on numercial columns
num_columns=df.select_dtypes(include=[np.number]).columns
num_columns
```

```python
#Standard Scaling
scaler=StandardScaler()
df_standard_scaled=df_clean.copy()
df_standard_scaled[num_columns]=scaler.fit_transform(df_clean[num_columns])
df_standard_scaled
```

```python
#MinMax Scaling
minmax_scaler=MinMaxScaler()
df_minmax_scaled=df_clean.copy()
df_minmax_scaled[num_columns]=minmax_scaler.fit_transform(df_clean[num_columns])
df_minmax_scaled
```

```python
#Summary
summary="""
Scaling Summary
1. Standard Scaling (z-score Normalization):
-Formula: (x - mean) / standard deviation
-Centers data around mean = 0 and standard deviation = 1
-Best for algorithms assuming normally distributed data (e.g., linear regression, SVM, PCA)
-Helps when data has outliers, as it reduces their relative impact compared to Min-Max scaling.

2. Min-Max Scaling (Normalization):
-Formula: (x - min) /(max - min)
-Transforms features into range (0,1)
-Preferred for distance-based algorithms like, KNN, Neural Networks, and Clustering where features must be comparable in scales
-Sensitive to outliers since extreme values can compress most data points into a small range.


In this dataset, features such as age, capital gain, hours_per_week can be effectively scaled using StandardScaler for most ML models.
However, MinMax Scaler is useful for neural networks where gradient-based optimization benefits from bounded input ranges.
"""
print(summary)

```

```python
#Encoding Techniques
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df=pd.read_csv('adult_with_headers.csv')
df
```

```python
categorical_columns=df.select_dtypes(include=['object']).columns
categorical_columns
```

```python
low_cardinality_columns=[col for col in categorical_columns if df[col].nunique() <= 5]
low_cardinality_columns
```

```python
#Applying One-Hot Encoding to low cardinality columns
df_onehot=pd.get_dummies(df, columns=low_cardinality_columns)
df_onehot
```

```python
high_cardinality_columns=[col for col in categorical_columns if df[col].nunique() > 5]
high_cardinality_columns
```

```python
#Applying Label Encoder to high cardinality columns
df_encoded=df.copy()
label_mappings={}
for col in high_cardinality_columns:
  #convert to string and fill missing values with a sentinel
  df_encoded[col]= df_encoded[col].astype(str).fillna('missing')
  le=LabelEncoder()
  df_encoded[col]=le.fit_transform(df_encoded[col])
  label_mappings[col]=le.classes_
  df_encoded
  print(df_encoded)
```

```python
#Discussion on Pros and Cons
Encoding= """
Encoding:

1.One-Hot Encoding:
 -Converts categorical values into multiple binary columns (0 or 1).
 -Example: 'Sex' => 'Sex Male', 'Sex Female'.
 -**Pros:**
       -No ordinal relationship implied (safe for non-ordered categories).
       -Works well for algorithms like Logical Regression and Tree-based models.
 -**Cons:**
       -Increases dimensionality (especially for high-cardinality columns).
       -Can cause sparse matrices, leading to higher memory usuage.

2.Label Encoding:
 -Assigns each category to an integer label (e.g., Red=0, Blue=1, Green=2).
 -**Pros:**
        -Simple and memory efficient.
        -Preserves all information in single column.
 -**Cons:**
        -Implies an artifical ordinal relationship between categories (which may mislead models).
        -Not ideal for linear models but accepted for Tree-based algorithms (Decision Trees, Random Forests).

"""
print(Encoding)
```

```python
#Feature Engineering
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('adult_with_headers.csv')
df
```

```python
num_columns=df.select_dtypes(include=[np.number]).columns
num_columns
```

```python
skewness=df[num_columns].skew().sort_values()
skewness
```

```python
highly_skewed=skewness[abs(skewness) > 1]
highly_skewed
```

```python
#Applying Log Transformation to one skewed feature (capital_gain)
df['capital_gain']=np.log1p(df['capital_gain'])
df['capital_gain']
```

```python
#Creating Two New Features
#Feature 1: Age Group (categorical grouping of age)
df['age group'] =pd.cut(df['age'], bins=[0, 25, 45, 65, 100], labels=['young', 'middle_aged', 'senior', 'old'])
df['age group']
```

```python
#Feature 2: Work hours category (based on weekly hours)
df['work_hours_category']=pd.cut(df['hours_per_week'], bins=[0, 25, 45, 65, 100], labels=['Part-time', 'Full-time', 'Overtime', 'Extreme'])
df['work_hours_category']
```

```python
#New features and Skewness Report
summary=pd.DataFrame({
    'Feature': skewness.index,
    'skewness': skewness.values,
    'Transformation_Required': ['Log' if f == 'capital_gain' else 'None' for  f in skewness.index]
})
print("Skewness and Transformation summary", summary)
```

```python
#Explanation for Feature Engineering
Explanation= """
Feature Engineering Rationale:


1. **New Feature 1: age group**
                  -Rationale: The effect of age on income is often nonlinear. Grouping age into catergories helps models
                   capture age based income differences (e.g., younger individuals might be students, middle-age individuals in peak careers,
                   old individuals near retirement).

2. **New Feature 2: work_hours_category**
                  -Rationale: Income potential and job type often relate to weekly work hours. Categorizing hours_per_week
                  helps model detect patterns between part-time, full-time, overtime workers.

3. **Transformation: capital_gain_log**
                  -Reason: 'capital_gain' is highly right skewed (most people have zero gain , few have very high values).
                  Applying log transformation normalizes it's distributio, reducing the influence of extreme outliers and
                  improving model performance for linear and distance-based algorithms.

These transformations improve model interpretabilty and roubstness by reducing skewness and capturing important behavioral
patterns from raw numeric data.

"""
print(Explanation)
```

```python
#Feature Selection
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('adult_with_headers.csv')
df
```

```python
num_columns= df.select_dtypes(include=['int64', 'float64']).columns
num_columns
```

```python
#Applying IsolationForest for outlier detection
iso=IsolationForest(contamination=0.02, random_state=42)
outlier_pred= iso.fit_predict(df[num_columns])
df['outlier']=outlier_pred
df['outlier']
```

```python
outlier_count= (df['outlier'] == -1).sum()
outlier_count
```

```python
df_clean=df[df['outlier'] == 1].drop(columns=['outlier'])
df_clean
```

```python
#Correlation matrix for numerical features (since PPS not available)
correlation_matrix=df_clean[num_columns].corr()
correlation_matrix
```

```python
#Discussion
Discussion= f"""
Outlier detection and Correlation Analysis:

1. **IsolationForest for Outlier Removal:**
   -IsolationForest identified {outlier_count} outliers approximately 2% of the dataset).
   -Outliers are data points that deviate significantly from the rest of the data distribution.
   -Removng these helps the model generalize better, prevents overfitting, and stbilizer parameters estimation.
   -However, excessive outlier removal may discard rare but valuable cases (e.g., 'high-income individuals').

   **Impact of Outliers:**
   -Outliers can disort model coefficients in linear model and inflate error metrics.
   -They also impact scaling, as min-max scaling and normalization are sensitve to extreme values.
   -Tree-based models are generally more robust but may still misallocate splits.

2. **Correlation Matrix:**
   -Correlation measures linear relationships between numeric features.
   -In the dataset, features like "capital_gain" and "hours_per_week" may show moderate positive correlation with "income".
   -Correlation helps identify related numeric variables but fails to detect non-linear dependcies.

3. As PPS feature was not available the results of the features are not shown and the relationship is also not mentioned


***Conclusion:***
             -Outlier removal via IsolationForest enhances data quality and reduces noise.
             -Correlation reveals direct numeric association but misses complex relationships.


"""
print(Discussion)
```
